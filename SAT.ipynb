```python
#13 Created a standard deviation function using list comprehensions
import numpy as np
import math
def standard_deviation(column):
    mean_value = np.mean(column) # calculate mean of column
    differences = [(i-mean_value)**2 for i in column] # go through each index in column and subtract that value from the mean value and square it 
    summed_differences = sum(differences) # sum the differences
    avg_differences = summed_differences / (len(column) - 1) # divide the sum by N
    std_dev = math.sqrt(avg_differences) #taking the square root 
    return std_dev
```

```python
#13
print 'The standard deviation of Rate is', standard_deviation(pand_sat_scores['Rate'])
print 'The standard deviation of Verbal is',standard_deviation(pand_sat_scores['Verbal'])
print 'The standard deviation of Math is', standard_deviation(pand_sat_scores['Math'])

```

20.

From the graph there is is an obvious negative correlation between the rate in which the students take the SATs and what their Math and Verbal score are. This is also backed up by the correlation coefficients of -.77 for Rate and Math and -.89 for Rate and Verbal. 

Students take either the SAT or ACT depending on where they are located within the U.S. The southern and central states have lower rates of students taking the SATs as they more predominantly take the ACTs. The fewer students who do take the SATs in those areas appear to be the more intelligent students and therefore weigh the mean score higher. They could possibly be taking the SATs to enroll in more prestigious schools or schools not in their area. When you get to the states that have around a 50% rate of taking the test, the population is large enough that the negative correlation between rate and test score ends.

By seperating the rates by greater than 50 and less than 50 and re-evaluating the distributions, they become more aligned with what we expect from a normal distribution. 

```python
# creating new database with SAT scores greater than 50
sat_50 = pand_sat_scores[pand_sat_scores.Rate >= 50]

import seaborn as sns
sns.regplot(y= sat_50.Verbal, x=sat_50.Rate, label = 'Verbal')
plt.legend(loc="best")
v=sns.regplot(y= sat_50.Math, x=sat_50.Rate, label= 'Math')
plt.legend(loc="best")
plt.ylabel('Math and Verbal')
plt.title('Rate greater than 50')

#testing for correlations
import scipy.stats as sp
print 'Correlation between Rate >= 50 and Math:', sp.stats.pearsonr(sat_50.Rate, sat_50.Math)[0]
print 'Correlation between Verbal >= 50 and Verbal:', sp.stats.pearsonr(sat_50.Rate, sat_50.Verbal)[0]
print 'Correlation between Verbal and Math with Rate >= 50:', sp.stats.pearsonr(sat_50.Math, sat_50.Verbal)[0]
```

```python
# create new database for SAT scores < 50
sat_less = pand_sat_scores[pand_sat_scores.Rate < 50]
sns.regplot(y= sat_less.Verbal, x=sat_less.Rate, label = 'Verbal')
plt.legend(loc="best")
v=sns.regplot(y= sat_less.Math, x=sat_less.Rate, label= 'Math')
plt.legend(loc="best")
plt.ylabel('Math and Verbal')
plt.title('Rate less than 50')

print 'Correlation between Rate < 50 and Math:', sp.stats.pearsonr(sat_less.Rate, sat_less.Math)[0]
print 'Correlation between Verbal < 50 and Verbal:', sp.stats.pearsonr(sat_less.Rate, sat_less.Verbal)[0]
print 'Correlation between Verbal and Math with Rate < 50:', sp.stats.pearsonr(sat_less.Math, sat_less.Verbal)[0]
```
